# 전국 노스페이스 매장 분석


### 요약
- 아웃도어 의류 브랜드 "노스페이스"는 주로 인구가 많은 도심지역(서울, 경기, 부산)등에 분포한다.
- 반대로 주로 등산이나 캠핑을 할 수 있는 산간지역(강원도, 전라도)은 매장 수가 작다.
- 아웃도어가 발생하는 지역이 아닌 인구 분포와 상관관계가 높을 것으로 판단되며 정확한 데이터를 위해 인구 데이터를 제시할 필요가 있음
- 추후 아웃도어 의류 브랜드들의 입점한 데이터를 비교해 볼 필요가 있음

```python
# 모듈 사용
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

# HTML 가져오기
url = 'https://www.thenorthfacekorea.co.kr/store?_find=&_search=state&_condition=like&_find=&_search=name&_condition=like&northface=%EB%85%B8%EC%8A%A4%ED%8E%98%EC%9D%B4%EC%8A%A4&pageSize=100&_find=%EB%85%B8%EC%8A%A4%ED%8E%98%EC%9D%B4%EC%8A%A4&_search=brandName&_condition=like'
page = req.urlopen(url)
soup = BeautifulSoup(page,'html.parser')
```

```python
# 노스페이스 매장 이름 추출하기
north_lists=soup.select('ul > li > a > h2')
count = 0
indexLists = []  # index
titleLists = []  # 점포 이름

for north in north_lists:
    count += 1
    print(count, " : ", north.string)
    indexLists.append(count)
    titleLists.append(north.string)
```
